```python
def range_w_b(data):
    min_x = data[0,0]
    min_y = data[0,1]
    max_y = data[0,1]
    max_x = data[0,0]
    for i in data:
        x = i[0]
        y = i[1]
        if x < min_x:
            min_x = x
        if x > max_x:
            max_x = x
        if y < min_y:  
            min_y = y
        if y > max_y:
            max_y = y
    return min_x, min_y, max_x, max_y
```
